```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv('Amazon Sale Report.csv', encoding='unicode_escape', low_memory=False) #low_memory=False tells Pandas to read the whole file at once
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop(columns = ['ASIN', 'SKU', 'B2B', 'fulfilled-by'], inplace = True)
```

```python
df.dropna(inplace = True)
```

```python
df.info()
```

```python
df.columns
```

```python
df.describe()
```

```python
# Top selling products with their counts

ax = sns.countplot( x = 'Category', data = df, )
sns.set(rc={'figure.figsize':(10,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['ship-state'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending = False).head(10)
sns.set(rc = {'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'ship-state' , y = 'Amount', hue = 'ship-state')
```

```python
avg_order_val = np.mean(df['Amount'])
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Amount', y='ship-country')
plt.title('Order Value Distribution India')
```

```python
# Calculate average order value
overall_aov = df['Amount'].mean()
print(f"Overall Average Order Value: rs {overall_aov:.2f}")
```

```python
# Calculate total revenue generate
total_revenue = df['Amount'].sum()
print(f'Total revenue generated : rs {total_revenue:.2f}')
```

```python
category_revenue = df.groupby('Category', as_index=False)['Amount'].sum()
sns.barplot( x = 'Category', y = 'Amount', data = category_revenue, hue = 'Category')
sns.set(rc = {'figure.figsize':(15,5)})
plt.title('Total Revenue by Category')
plt.xticks(rotation=45)
```

```python
plt.hist(df['Amount'], bins=20, color='lightgreen', edgecolor='black')
```

```python
df['Date'] = pd.to_datetime(df['Date'])
rev = df.groupby('Date', as_index=False)['Amount'].sum()
sns.lineplot(data=rev, x='Date', y='Amount')
```

```python
qty_vs_amt = df.groupby('Qty', as_index = False)['Amount'].sum()
sns.scatterplot( x = 'Qty', y = 'Amount', data = qty_vs_amt)
```

```python
sns.boxplot( x = 'Category', hue = 'Courier Status',y = 'Size', data=df)
```
